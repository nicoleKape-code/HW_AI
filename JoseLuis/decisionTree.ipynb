```python
state = 0 
while state != 80:
    if state == 0:
        print("Hola viajero, bienvenido a mi tienda")
        print("Vendo pociones (1) y espadas (2) ") 
        print("Deseas repetir el diálogo '0'?")
        choice = int(input("Selecciona la opción: "))
        if choice == 1:
            state = 1
        elif choice == 2: 
            state = 2
        else: 
            state = 0
    if state == 1: 
        print("Veo que te interesan las pociones, vendemos pociones de chocolate (1) o de cacahuate (2)")
        choice = int(input("Selecciona la opción: "))
        if choice == 1:
            state = 3
        elif choice == 2: 
            state = 4
        else: 
            print("No te entendí")
        
        if state == 3:
            print("Veo que te interesan las pociones de chocolate, escoges vender (1) o salir (2)")
            choice = int(input("Selecciona la opción: "))

            if choice == 1:
                state = 8
            elif choice == 2: 
                state = 7
        
        elif state == 4:
            print("Veo que te interesan las pociones de cacahuate, escoges vender (1) o salir (2)")
            choice = int(input("Selecciona la opción: "))

            if choice == 1:
                state = 8
            elif choice == 2: 
                state = 7

    elif state == 2:
        print("Veo que te interesan las espadas, vendemos dagas (1) o dientes de león (2)")
        choice = int(input("Selecciona la opción: "))
        if choice == 1:
            state = 5
        elif choice == 2: 
            state = 6
        else: 
            print("No te entendí")

        
        if state == 5:
            print("Veo que te interesan las dagas, escoges vender (1) o salir (2)")
            choice = int(input("Selecciona la opción: "))

            if choice == 1:
                state = 8
            elif choice == 2: 
                state = 7
        
        elif state == 6:
            print("Veo que te interesan los dientes de leon, escoges vender (1) o salir (2)")
            choice = int(input("Selecciona la opción: "))

            if choice == 1:
                state = 8
            elif choice == 2: 
                state = 7

    if state == 8: 
        print("Este es el fin")
    elif state == 7: 
        print("Este es el fin")



    
```
